```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### data preprocessing 

```python
raw_data = pd.read_csv("../../dataset/dataset.csv");
raw_data.head()
```

```python
X = raw_data["Head Size(cm^3)"]#fetching X column and converting to numpy array 
X = np.array(X)
X = X.reshape(X.shape[0],1)
X.shape
```

```python
X = X/np.max(X,axis=0) #normalizing data
X.shape
```

```python
Y = raw_data["Brain Weight(grams)"]#fetching X column and converting to numpy array 
Y = np.array(Y)
Y = Y.reshape(Y.shape[0],1)
Y.shape
```

```python
Y = Y/np.max(Y,axis=0) #normailizind data
```

```python
#performing gradient descent for single Linear Regression
alpha = 0.01 #this it the learning rate(step-size)
epsilon = 0.0000001 #this is the error tolerance . alpha and epsilona re hyperparameters

Q0_inital = np.random.randn(1,1) 
Q1__intial = np.random.randn(1,1)#these parameters can be in vector form also
print(Q0_inital,Q1__intial)
```

```python

x = X.T
m = len(X)
i = 0
iterations = []#this is to check the iteration of loop
loss = []
#cost_function_dels = np.dot(dels.T,)
while (True):
    #calculating the cost function derivative for gradient descent 
    dels = (((Q0_inital + (Q1__intial*X)) - Y))
    c = np.dot(x,dels)
    cost_function_theta0 = np.mean(dels)
    cost_function_theta1 = np.mean(c)
    
    #putting values of derivative in gradient descent
    Q0_new = Q0_inital - (alpha * cost_function_theta0)
    Q1_new = Q1__intial - (alpha * cost_function_theta1)
    
    function_new =  np.sum(((Q0_new + (Q1_new*X)) - Y)**2)/(m*2)
    function_old =  np.sum(((Q0_inital + (Q1__intial*X)) - Y)**2)/(m*2)
    
    if abs(function_new-function_old)<epsilon:
        break
    i += 1
    iterations.append(i)
    loss.append(function_old)
    print("The cost function {} in iteration {}".format(function_old,i))
    Q0_inital = Q0_new
    Q1__intial = Q1_new
```

```python

plt.scatter(iterations,loss)
plt.xlabel("iterations")
plt.ylabel("loss")
```
