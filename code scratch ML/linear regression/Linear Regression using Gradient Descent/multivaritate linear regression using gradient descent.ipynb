```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data_training = pd.read_csv("../../dataset/Boston/train.csv");
raw_data_testing = pd.read_csv("../../dataset/Boston/test.csv");
raw_data_training.head()
```

```python
#performing data preprocessing

end = (raw_data_training.shape[1])-1
X = raw_data_training.iloc[:,1:end]#selcting the data using iloc columns iloc[:,0:n-1]
Y = raw_data_training.iloc[:,end:end+1] #selcting the last column for actual y  print
print("X:{}\nY:{}".format(X.shape,Y.shape))
```

```python
#converting dataFrame into numpy array
X = np.array(X) 
Y = np.array(Y)
print("X:{}\nY:{}".format(X.shape,Y.shape))
```

```python
#permorming Feature scaling for gradient descent
X = X/np.max(X,axis=0)
Y = Y/np.max(Y,axis=0)
print("X:{}\nY:{}".format(X.shape,Y.shape))
```

```python
#performing gradient descent for single Linear Regression
alpha = 0.001 #this it the learning rate(step-size)
epsilon = 0.0000001 #this is the error tolerance . alpha and epsilona re hyperparameters
n = X.shape[0]
Q0_inital = np.random.randn(1,1) 
Q1__intial = np.random.randn(n,1)#these parameters can be in vector form also
#print(Q0_inital,"\n \n \n", Q1__intial)
```

```python

x = X.T
m = len(X)
i = 0
iterations = []#this is to check the iteration of loop
loss = []
#cost_function_dels = np.dot(dels.T,)
while (True):
    #calculating the cost function derivative for gradient descent 
    dels = (((Q0_inital + (Q1__intial*X)) - Y))
    c = np.dot(x,dels)
    cost_function_theta0 = np.mean(dels)
    cost_function_theta1 = np.mean(c)
    
    #putting values of derivative in gradient descent
    Q0_new = Q0_inital - (alpha * cost_function_theta0)
    Q1_new = Q1__intial - (alpha * cost_function_theta1)
    
    function_new =  np.sum(((Q0_new + (Q1_new*X)) - Y)**2)/(m*2)
    function_old =  np.sum(((Q0_inital + (Q1__intial*X)) - Y)**2)/(m*2)
    if abs(function_new-function_old)<epsilon:
        break
    i += 1
    iterations.append(i)
    loss.append(function_old)
    print("The cost function {} in iteration {}".format(function_old,i))
    Q0_inital = Q0_new
    Q1__intial = Q1_new
```

```python
print(Q0_new,"\n \n \n",Q1_new)
```

```python
print(iterations[-1])
```

```python
plt.scatter(iterations,loss)
plt.xlabel("iterations")
plt.ylabel("loss")
```

```python
#data preprocessing for testing data
end1 = (raw_data_testing.shape[1])-1
X1 = raw_data_testing.iloc[:,1:end1]#selcting the data using iloc columns iloc[:,0:n-1]
X1 = np.array(X1) 
X1 = X1/np.max(X1,axis=0)
print("X1:{}".format(X1.shape))
```

```python
x_mean = np.mean(X,axis=1)
x_mean = x_mean.reshape(x_mean.shape[0],1)
print(Y.shape,x_mean.shape)
x_mean_test = np.mean(X1,axis=1) 
x_mean_test = x_mean_test.reshape(X1.shape[0],1)
#performing the visalization using Q0new and Q1new
y_pred = Q0_new + Q1_new * x_mean
print(x_mean_test.shape)
```
